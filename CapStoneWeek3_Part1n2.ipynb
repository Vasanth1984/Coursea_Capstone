### Scrape data from website

```python
d = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
df = d[0]
print(df.shape)
df.head()
```

#### Clean scraped data 

```python
A=[]

#convert scraped dataframe to a list
A.append(df[1])
A = df.values.tolist()
print(A)

#convert nested list to a flat list
flat_list = []
for sublist in A:
    for item in sublist:
        flat_list.append(item)

#print(flat_list)

#converted flat list to a dataframe
df_n = pd.DataFrame(flat_list) 
df_n.shape
```

```python
df_n[1] = df_n[0]
df_n[2] = df_n[0]
df_n
```

```python
#Only retain first 3 characters to show postal code
df_n[0] = df_n[0].astype(str).str[:3]
df_n[0]
```

```python
#Drop all characters after the ''(' and the first 3 characters to show Borough
df_n[1] = df_n[1].str.split('(').str[0]
df_n[1]

df_n[1] = df_n[1].astype(str).str[3:]
df_n[1]
```

```python
#Drop all characters before the ''(' to show neighborhood
df_n[2] = df_n[2].str.split('(').str[1]
df_n[2]
```

```python
#Name the columns
df_n.columns = ['PostalCode', 'Borough', 'Neighborhood']
df_n
```

```python
#remove ')' in Neighborhood column
df_n['Neighborhood']=df_n['Neighborhood'].str.replace(")", "")

df_n
```

```python
# Get names of indexes for which column Borough is not assigned
index = df_n[df_n['Borough'] == 'Not assigned' ].index
# Delete these row indexes from dataFrame
df_n.drop(index , inplace=True)

df_n
```

```python
len(df_n['PostalCode'].unique())
#Only unique postal code exists
```

```python
#rplaceing '/' with a ','
df_n['Neighborhood']=df_n['Neighborhood'].str.replace("/", ",")
df_n['Borough']=df_n['Borough'].str.replace("/", ",")
df_n
```

```python
#Replace all Nan values in Neighborhood with adjacent values in Borough 
df_n.loc[(pd.isnull(df_n.Neighborhood)), 'Neighborhood'] = df_n.Borough
df_n
```

```python
#Reset Index
df_n.reset_index(drop=True, inplace=True)
df_n
```
